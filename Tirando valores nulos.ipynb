## Tratamento de dados faltantes - Condicional 

```python
# Usado em casos onde não faz sentido descartar os dados mesmo com dados nulos, mas pode ter situações onde não faz sentido
# Que tenha os dados nulos, por exemplo: se um apartamento tiver o condomínio nulo, deve ser eliminado. Pois no nosso entendimento
# todo apartamento deve ter um condomínio

selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())

dados[selecao].shape[0]

dados = dados[~selecao]

dados[dados['Condominio'].isnull()].shape[0]

# Inserindo valores, em áreas nulas
dados.fillna(0, inplace=True)

# Criar novo dataframe setando em cada coluna específica o tipo de valor que deseja
dados = dados.fillna({ 'Condominio': 0, 'IPTU': 0 })
```
