# Relatório de Análise III

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=";")
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
]
```

```python
residencial
```

```python
# Retorna a correspondência em booleano dos dados contidos no dataframe
selecao = dados['Tipo'].isin(residencial)
```

```python
# Seleção de dataframe: ao passar o tipo de informação
# será mantidos os dados que estão como True, descartando a testagem que retor,a false 
dados_residencial = dados[selecao]

dados_range_index = range(len(dados_residencial))

dados_residencial.index = dados_range_index

dados_residencial.columns.name = 'Id'

dados_residencial
```

## Exportando base de dados

```python
dados_residencial.to_csv('dados/alguel_dados_residencial.csv', sep=';', index=False)

# Index false fará com que não exporte com o index, 
# na hora de importar novamente o pandas automaticamente vai criar novamente os indexes
```
